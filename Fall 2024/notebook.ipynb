```python
import numpy as np
import matplotlib.pyplot as plt
import networkx as nx
import random
import kmax as kx
import utility as util
```

# Mean-Field Last-Passage Percolation
### Test Setup

```python
n = 20
G = nx.complete_graph(n)

# Add random edge weights
for e in G.edges:
    G[e[0]][e[1]]["weight"] = random.uniform(0,1)

kmax = kx.generate_kmax(G, 2)
```

### Drawing

```python
plt.figure(1)
kx.custom_draw(kmax, "temp/kmax.png", layout=nx.spring_layout, draw_edge_weights=False)

plt.figure(2)
kx.custom_draw(G, "temp/G.png", draw_edge_weights=False)
```

### Misc.

```python
# Export to GraphML
nx.write_graphml(kmax, "temp/kmax.xml")
```
